```python
import json
import pandas as pd
```

```python
with open('logs/mints.json', 'r', encoding='utf-8') as file:
    mints = json.load(file)
```

```python
with open('logs/burns.json', 'r', encoding='utf-8') as file:
    burns = json.load(file)
```

```python
with open('logs/swaps.json', 'r', encoding='utf-8') as file:
    swaps = json.load(file)
```

```python
mints = pd.DataFrame(mints)
burns = pd.DataFrame(burns)
swaps = pd.DataFrame(swaps)
mints['event'] = 'Mint'
burns['event'] = 'Burn'
swaps['event'] = 'Swap'
```

```python
logs = pd.concat([mints, burns, swaps], ignore_index=True).fillna(0)
logs['logIndex'] = logs['logIndex'].astype(int)
logs['blockNumber'] = logs['blockNumber'].astype(int)
logs = logs.sort_values(by=['blockNumber', 'logIndex'], ascending=[True, True])
logs.to_csv('logs.csv', header=True, index=False)
logs
```
